## DATA VISUALIZATION

### Most paid apps vs their installs

```python
plt.barh(top_paid['App'],top_paid['Installs'])
plt.title('Most paid apps vs their installs',fontsize=15,color='red')
ax=plt.axes()
ax.set_facecolor('orange')
plt.ylabel('Apps',fontsize=15)
plt.xlabel('Installs',fontsize=15)
plt.grid(True)
plt.show()
```

### Most Rated apps vs their Reviews

```python
plt.barh(m_rated['App'],m_rated['Reviews'])
plt.title('Most rated apps vs their reviews',fontsize=15,color='red')
ax=plt.axes()
ax.set_facecolor('yellow')
plt.ylabel('Apps',fontsize=15)
plt.xlabel('Installs',fontsize=15)
plt.grid(True)
plt.show()
```

### Most rated Apps as per category

```python
sns.set_style('whitegrid')
sns.stripplot(x='Category',y='App',hue='Type',jitter=0.5,dodge=True,data=m_rated)
plt.show()
```

### Total Installs of all categories of various apps

```python
plt.figure(figsize=(5,10))
ax=plt.axes()
ax.set_facecolor('cyan')

plt.bar(apps_cat_top.Category,apps_cat_top.Installs,label='total installs',color='r',width=0.5)

plt.legend()
plt.xlabel('Category')
plt.ylabel('Installs')
plt.title('Top 10 most Installed apps')
plt.xticks(rotation=70)
plt.show()
```

### Reviews and Category of most Unrated app

```python
y=u_rated.Reviews
l=u_rated.Category
c=['r','b','g','k','y']

plt.pie(y,labels=l,startangle=180,shadow=True,colors=c,explode=(0,0,0,0.5,0,0,0,0,0,0))
plt.legend(loc='upper left')
plt.title('Most Unrated Apps')
plt.show()
```
